# Sales Analysis of EV Cars
* Analyze the sales of top 10 car models 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
```

```python
models=["Tesla Model3","Tesla ModelS","Nissan Leaf","Tesla ModelX","ChevyBolt","BMWi3","Other"]
total_sales=[296392,162285,141888,87033,58208,41988,51202]
colors=["blue","cornflowerblue","yellow","lightskyblue","red","orange","pink"]
explode = (0,0,0,0,0,0,0.2)
```

```python
plt.pie(total_sales,labels=models,colors=colors,explode=explode,autopct="%1.0f%%",labeldistance=1.2,startangle=140)
plt.axis("equal")
plt.show
```
